```python
#creating a tuple
my_tuple = (1, 2, 3, 4, 5)
print(my_tuple) # Output: (1, 2, 3, 4, 5)
print(type(my_tuple)) # Output: <class 'tuple'>
```

```python
#accessing elements in a tuple
print(my_tuple[0]) # Output: 1
print(my_tuple[1]) # Output: 2
print(my_tuple[-1]) # Output: 5
```

```python
#slicing a tuple
print(my_tuple[1:4]) # Output: (2, 3, 4) creates a new tuple with the elements from index 1 to 3
print(my_tuple[:3]) # Output: (1, 2, 3) creates a new tuple with the first three elements
print(my_tuple[3:]) # Output: (4, 5) creates a new tuple with the elements from index 3 to the end
```

```python
#tuple operations
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
print(tuple1 + tuple2) # Output: (1, 2, 3, 4, 5, 6) concatenates two tuples
print(tuple1 * 2) # Output: (1, 2, 3, 1, 2, 3) creates a new tuple by repeating the elements of tuple1 twice
```

```python
#packing and unpacking tuples
packed_tuple = 1, "hello", 3.14
print(packed_tuple) # Output: (1, 'hello', 3.14)
a, b, c = packed_tuple
print(a) # Output: 1
print(b) # Output: hello
print(c) # Output: 3.14
```
