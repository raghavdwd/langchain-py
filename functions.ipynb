```python
def even_or_odd(number):
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"
```

```python
even_or_odd(10)
```

```python
#variable lenght arguments
# *args for non-keyword variable length arguments, we can also use *raghav in place of *args
def sum_all(*args):
    return sum(args)
```

```python
sum_all(2,4,5,6,6,4,0)
```

```python
#keywords arguments
def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

```python
print_details(name="Alice", age=30, city="New York")
```

```python
# we can also use positional arguments along with keyword arguments
def print_info(*args, **kwargs):
    for key in args:
        print(f"Positional argument: {key}")
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

```python
print_info("raghav",123, 432, name="Alice", age=30, city="New York")
```

```python
def convert_temperature(value, unit):
    if unit == "C":
        return (value * 9/5) + 32
    elif unit == "F":
        return (value - 32) * 5/9
    else:
        return "Invalid unit"
```

```python
print(convert_temperature(100, "C"))
print(convert_temperature(212, "F"))
```

```python
def is_strong_password(password):
    if len(password) < 8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    return True

# if not any means ..... if there is no digit in the password then it will return False and if there is at least one digit then it will return True and same for uppercase and lowercase letters.
```

```python
print(is_strong_password("Password123"))
print(is_strong_password("pass"))
```

```python
def calculate_total_cost(cart):
    total_cost = 0
    for item in cart:
        total_cost += item['price'] * item['quantity']
    return total_cost
```

```python
cart = [
    {"name": "Apple", "price": 1.2, "quantity": 5},
    {"name": "Banana", "price": 0.8, "quantity": 3}
]
print(calculate_total_cost(cart))
```

```python
def isPalindrome(string):
    s = string.replace(" ", "").lower()
    return s == s[::-1] # checking if the string is equal to its reverse ::-1 means string slicing to reverse the string
```

```python
print(isPalindrome("A man a plan a canal Panama"))
print(isPalindrome("Hello"))
```

```python
def get_factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * get_factorial(n - 1)
    
```

```python
print(get_factorial(5))
```
