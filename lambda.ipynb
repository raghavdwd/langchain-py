```python
#lambda functions: these functions are defined without a name and are often used for short-term operations. They are created using the lambda keyword and can take any number of arguments but can only have one expression. The syntax is: lambda arguments: expression. For example, to create a function that adds two numbers, you can use: add = lambda x, y: x + y. You can then call this function like this: print(add(2, 3)), which will output 5. Lambda functions are commonly used in situations where a simple function is needed for a short period of time, such as in sorting or filtering data.
```

```python
addition = lambda a,b: a + b
print(type(addition)) # Output: <class 'function'>
print(addition(5, 3)) # Output: 8
```

```python
isEven = lambda x: x % 2 == 0
print(isEven(4)) # Output: True
print(isEven(5)) # Output: False
```

```python
add_all = lambda *args: sum(args)
print(add_all(1, 2, 3, 4)) # Output: 10
```

```python
# map(): The map() function applies a given function to each item of an iterable (like a list) and returns a map object (which is an iterator). The syntax is: map(function, iterable). For example, if you have a list of numbers and want to square each number, you can use: numbers = [1, 2, 3, 4]; squared = map(lambda x: x**2, numbers); print(list(squared)), which will output [1, 4, 9, 16].
```

```python
nums = [1, 2, 3, 4, 5]
squared_nums = map(lambda x: x**2, nums)
print(list(squared_nums)) # Output: [1, 4, 9, 16, 25]
```

```python
#map multiple iterables: The map() function can also take multiple iterables as arguments. In this case, the function will be applied to the items of the iterables in parallel. The syntax is: map(function, iterable1, iterable2, ...). For example, if you have two lists of numbers and want to add them together element-wise, you can use: list1 = [1, 2, 3]; list2 = [4, 5, 6]; result = map(lambda x, y: x + y, list1, list2); print(list(result)), which will output [5, 7, 9].
```

```python
list1 = [1, 2, 3]
list2 = [5,5,6,5]
result = map(lambda x, y: x + y, list1, list2)
print(list(result)) # Output: [6, 7, 9]
```

```python
words = ['raghav','saumya', 'tiwari']
res = list(map(str.upper, words))
print(res)
```

```python
# filter(): The filter() function constructs an iterator from elements of an iterable for which a function returns true. The syntax is: filter(function, iterable). For example, if you have a list of numbers and want to filter out the even numbers, you can use: numbers = [1, 2, 3, 4, 5]; even_numbers = filter(lambda x: x % 2 == 0, numbers); print(list(even_numbers)), which will output [2, 4].
```

```python
nums = [1, 2, 3, 4, 5]
greater_than_3 = filter(lambda x: x > 3, nums)
print(list(greater_than_3)) # Output: [4, 5]
```
