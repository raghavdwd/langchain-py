```python
with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
#read line by line
with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip()) #strip() is used to remove the newline character at the end of each line
```

```python
# writing a file(overwriting)
with open('example.txt', 'w') as file:
    file.write("Hello world!! \n")
    file.write("This is raghav dwivedi \n")
```

```python
# without overwriting and by appending
with open('example.txt', 'a') as file:
    file.write("hello how are u \n");
    file.write("this is raghav \n")
```

```python
#writing a list of lines in a file
lines = ["htllo", "me and u", "raghaddwd"]
with open('example.txt', 'a') as file:
    file.writelines(lines)
```

```python
#writing into file then reading it:::
with open('example.txt', 'w+') as file:
    file.write("Hello world \n")
    file.write("This is ashu dwivedi \n")
    #move the file cursor to the beginning
    file.seek(0)
    #now read
    content = file.read()
    print(content)
```

```python
## counting lines, words and chars from a file
with open('example.txt', 'r') as file:
    lines = file.readlines()
    line_count = len(lines)
    word_count = sum(len(line.split()) for line in lines)
    char_count = sum(len(line) for line in lines)
    print(line_count)
    print(word_count)
    print(char_count)
```
